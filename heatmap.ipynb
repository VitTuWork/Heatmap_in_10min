```python
import pandas as pd # загружаем библиотек pandas
```

```python
# Загружаем таблицу из интернета
page_url = 'https://wciom.ru/index.php?id=236&uid=1140' # адрес страницы, где опубликована
wom_appreciate = pd.read_html(page_url, attrs = {'class': 'tablesorter'}) # в какой переменной будет хранится наша таблица(название), 'class': 'tablesorter' на этой странице, таблице присвоен такой класс, если не знаете, просто сотрите, будет так( pd.read_html(page_url))
```

```python
wom_appreciate = wom_appreciate[0] # убираем ненужное сверху
```

```python
wom_appreciate.columns = ['Ценности', 'Все опрошенные', '18-24 года','25-34 года','35-44 лет','45-59 лет','60 и старше'] # переименовываем столбцы
```

```python
wom_appreciate
```

```python
wom_appreciate = wom_appreciate.set_index('Ценности') # меняем числовой индекс строк на данные из столбца «Ценности»
```

```python
wom_appreciate
```

```python
# загружаем библиотеки Seaborn и Matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.subplots(figsize=(15,15)) # делаем размер таблицы побольше
fig = sns.heatmap(wom_appreciate, cmap="YlGnBu") # выводим нашу heatmap, cmap, это цветовая схема, какие еще бывают посмотрите в документации к Seaborn
```

```python
plt.subplots(figsize=(15,15))
fig = sns.heatmap(wom_appreciate, cmap="YlGnBu", linewidths=1) # добавим линии
```

```python
 plt.subplots(figsize=(15,15))
fig = sns.heatmap(wom_appreciate, cmap="YlGnBu", linewidths=1, fmt='d', annot_kws={"size": 10}, annot=True) # добавим числовые значения (fmt='d' – целые числа, fmt='d' – с дробной частью;  annot_kws – размер кегля)
```

```python
# сохраняем как хотим 
# fig.figure.savefig('heatmap.png', format='png' )
# fig.figure.savefig('heatmap.pdf', format='pdf' )
fig.figure.savefig('heatmap.svg', format='svg' )
```
